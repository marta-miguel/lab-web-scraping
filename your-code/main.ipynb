#### 1.1. Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
# your code here
response = requests.get(url)
#print(response)

soup = BeautifulSoup(response.content, "html.parser")

repos = soup.find_all("h1", attrs = {"class":"h3 lh-condensed"})
#repos

repo_list = list(map(lambda x: x.text.strip(), repos[0:]))
repo_list

cleaned_repos = []

for repo in repo_list:
    repo = repo.replace("/\n\n      ", " ")
    cleaned_repos.append(repo)

cleaned_repos
````

#### 2.2. Display the top 10 languages by number of native speakers stored in a pandas dataframe.
Hint: After finding the correct table you want to analyse, you can use a nested **for** loop to find the elements row by row (check out the 'td' and 'tr' tags). <br>An easier way to do it is using pd.read_html(), check out documentation [here](https://pandas.pydata.org/pandas-docs/version/0.23.4/generated/pandas.read_html.html).

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
# your code here
import pandas as pd

response = requests.get(url)
#print(response)
soup = BeautifulSoup(response.content, "html.parser")

## names of languages:

language = soup.find_all("a", attrs = {"class":"mw-redirect"})
languages = list(map(lambda x: x.text.strip(), language[7:17]))
languages

## number of native speakers:

number = soup.find_all("td")
numbers = list(map(lambda x: x.text.strip(), number[1:38]))
numbers
speakers = []

for i in range(0,len(numbers),4):
    speakers.append(numbers[i])
    
speakers

top10 = {}

for l,s in zip(languages, speakers):
    top10[l] = s
    
pd.DataFrame.from_dict(top10, orient = "index", columns = ["Number of Native Speakers"])
````

#### 3. Display IMDB's top 250 data (movie name, initial release, director name and stars) as a pandas dataframe.
Hint: If you hover over the title of the movie, you should see the director's name. Can you find where it's stored in the html?

````python
# This is the url you will scrape in this exercise 
url = 'https://www.imdb.com/chart/top'

response = requests.get(url)
print(response)
soup = BeautifulSoup(response.content, "html.parser")
````

````python
# # your code here

import re
import pandas as pd

# ## Movie titles

movies = soup.find_all("img")
titles = []

for name in movies:
    title = name.get("alt")
    titles.append(title)

titles

# ## Dates:

date = soup.find_all("span", attrs = {"class":"secondaryInfo"})
dates = list(map(lambda x: x.text.strip(), date[0:]))
dates

# ## Directors / Stars:

info = soup.find_all("a")
infos = []

for item in info:
    descr = item.get("title")
    infos.append(descr)

infos = infos[50:550]
description = []

for i in range(0, len(infos),2):
    description.append(infos[i])
    
descriptions = []

for line in description:
    descriptions.append(line.split("(dir.),"))

directors = []
stars = []

for line in descriptions:
    directors.append(line[0].strip())
    stars.append(line[1].strip())
    
directors
stars 

data = {"Movie title":titles, "Initial release":dates, "Director": directors, "Stars": stars}

top250 = pd.DataFrame(data)
top250.head(5)    
````
